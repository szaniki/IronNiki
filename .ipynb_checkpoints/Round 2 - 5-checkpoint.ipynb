## Get data

```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('/Users/szabonikolett/Desktop/Ironhack-Labs/lab-customer-analysis-round-2/files_for_lab/csv_files/marketing_customer_analysis.csv') 
pd.set_option('display.max_columns', None)  # to list all columns

display(dataset)
```

## Standardize header names

```python
#cols = []  
#for column in dataset.columns:
#    cols.append(column.lower())
#dataset.columns = cols
#dataset

cols = []
for column in range(len(dataset.columns)):
    cols.append(dataset.columns[column].lower().replace(' ', '_'))  # >> new best friend, replacing in column names space with _
dataset.columns = cols

print(dataset.columns)
```

## Numerical and categorical columns

## Check and deal with NaN values

```python
# Check the occurancy of NaN values / column

dataset.isna().sum() # number of nan values
round(dataset.isna().sum()/len(dataset),4)*100  # % of nan values
```

```python
# removing vehicle tpye column as half of it is missing, makes no sense to work with that much missing values

# already dropped the column, so code will not execute again 

dataset = dataset.drop(['vehicle type'], axis=1)
dataset
```

```python
# dropping 'unnamed' column as it has no relevance or added value, and appears to be = to the row number

dataset = dataset.drop(columns=['unnamed:_0'])

dataset
```

```python
# dropping duplicates 

dataset = dataset.drop_duplicates()
dataset 
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

y = numerical['Total Claim Amount']
X = numerical.drop(['Total Claim Amount'], axis=1) # axis=1 needed because we drop a column
lm = linear_model.LinearRegression()
lm.fit(X,y) 
print(lm.score(X,y))
y_pred = lm.predict(X)
print(mean_squared_error(y_pred,y))
```
